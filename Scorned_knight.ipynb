```python
import random
import time
import sys
def slow_print(text, delay=0.03):
    for char in text:
        sys.stdout.write(char)
        sys.stdout.flush()
        time.sleep(delay)
    print()
kill_count = 0
# Stats
#they all start at 1
class Stats:
 def __init__(self, strength=1, agility=1, defense=1, intelligence=1, constitution=1):
     self.strength = strength
     self.agility = agility
     self.defense = defense
     self.intelligence = intelligence
     self.constitution = constitution
#1 constitution increases health by 5
 def max_hp(self):
     return 50 + self.constitution * 5
#you can check your stats with this
 def __check_stats__(self):
     return (
         f"STR:{self.strength} "
         f"AGI:{self.agility} "
         f"DEF:{self.defense} "
         f"INT:{self.intelligence} "
         f"CON:{self.constitution}"
     )




# Character Class definition
class CharacterClass:
 def __init__(self, name, base_stats, allowed_weapon_types, description):
     self.name = name
     self.base_stats = base_stats
     self.allowed_weapon_types = allowed_weapon_types
     self.description = description




 def __repr__(self):
     return self.name



# Enemy Class definition(ALSO VERY IMPORTANT)
class EnemyClass:
 def __init__(self, name, base_stats, weapon, description, exp, min_hp, item_drop=None, weapon_drop=None,
              is_poisoned=False, poison_remaining_turns=0, poison_damage_on_turn=0,
              is_stunned=False, stun_remaining_turns=0,
              is_feared=False, fear_remaining_turns=0, feared_damage_reduction=0.0):
     self.name = name
     self.base_stats = base_stats
     self.weapon = weapon
     self.description = description
     self.exp = exp
     self.min_hp = min_hp
     self.item_drop = item_drop
     self.weapon_drop = weapon_drop

     # Status effects
     self.is_poisoned = is_poisoned
     self.poison_remaining_turns = poison_remaining_turns
     self.poison_damage_on_turn = poison_damage_on_turn
     self.is_stunned = is_stunned
     self.stun_remaining_turns = stun_remaining_turns
     self.is_feared = is_feared
     self.fear_remaining_turns = fear_remaining_turns
     self.feared_damage_reduction = feared_damage_reduction

 def max_hp(self):
     return self.min_hp + self.base_stats.constitution * 5

 def __repr__(self):
     return self.name
# Armor Class definition(ALSO VERY IMPORTANT)
class armor:
 def __init__(
     self, name, Armor_type, Protection, defense_requirement,
     allowed_classes=None, description="", allowed_enemies=None, armor_light=False, armor_heavy=False
 ):
     self.name = name
     self.Armor_type = Armor_type
     self.Protection = Protection
     self.defense_requirement = defense_requirement
     self.allowed_classes = allowed_classes if allowed_classes is not None else []
     self.allowed_enemies = allowed_enemies if allowed_enemies is not None else []
     self.description = description
     self.armor_light = armor_light
     self.armor_heavy = armor_heavy




 def __repr__(self):
     return self.name
# Items Class definition(ALSO VERY IMPORTANT)
class Items:
 def __init__(
     self, name, Is_Health_Pot=False, Is_Key=False, Is_Strength_pot=False, Is_Defense_pot=False, Is_Intelligence_pot=False, Is_speed_pot=False, Health_healed=None,
     Stat_Buff=None, allowed_classes=None, description="", allowed_enemies=None, light=False, heavy=False, Is_Book=False):
     self.name = name
     self.Is_Health_Pot = Is_Health_Pot
     self.Is_Strength_pot = Is_Strength_pot
     self.Is_Defense_pot = Is_Defense_pot
     self.Is_Intelligence_pot = Is_Intelligence_pot
     self.Is_speed_pot = Is_speed_pot
     self.Stat_Buff = Stat_Buff
     self.Is_Book = Is_Book
     self.Is_Key = Is_Key
     self.Health_healed = Health_healed
     self.allowed_classes = allowed_classes if allowed_classes is not None else []
     self.allowed_enemies = allowed_enemies if allowed_enemies is not None else []
     self.description = description




 def __repr__(self):
     return self.name
# Rooms Class definition(ALSO VERY IMPORTANT)
class Room:
 def __init__(
     self, name, North_connection=None, East_connection=None, South_connection=None, West_connection=None, description="", items=None, weapons=None, enemies=None, is_final=False, armor=None, is_boss_room=False
 ):
     self.name = name
     self.North_connection = North_connection
     self.East_connection = East_connection
     self.South_connection = South_connection
     self.West_connection = West_connection
     self.items = items if items is not None else []
     self.weapons = weapons if weapons is not None else []
     self.enemies = enemies if enemies is not None else []
     self.description = description
     self.is_final = is_final
     self.armor = armor if armor is not None else []
     self.is_boss_room = is_boss_room




 def __repr__(self):
     return self.name
# Weapon Class definition(ALSO VERY IMPORTANT)
class Weapon:
 def __init__(
     self, name, weapon_type, min_damage, max_damage, scaling_stats,
     allowed_classes=None, description="", dual_wield=False, light=False, heavy=False, allowed_enemies=None,
     has_poison=False, poison_chance=0.0, poison_duration=0, poison_damage_multiplier=0.0,
     has_stun=False, stun_chance=0.0, stun_duration=0,
     has_fear=False, fear_chance=0.0, fear_duration=0, fear_damage_reduction=0.0,
     has_lifesteal=False, lifesteal_base=0, lifesteal_per_con=0
 ):
     self.name = name
     self.weapon_type = weapon_type
     self.min_damage = min_damage
     self.max_damage = max_damage
     self.scaling_stats = scaling_stats
     self.allowed_classes = allowed_classes if allowed_classes is not None else []
     self.allowed_enemies = allowed_enemies if allowed_enemies is not None else []
     self.description = description
     self.dual_wield = dual_wield
     self.light = light
     self.heavy = heavy

     # New special effects attributes
     self.has_poison = has_poison
     self.poison_chance = poison_chance
     self.poison_duration = poison_duration
     self.poison_damage_multiplier = poison_damage_multiplier
     self.has_stun = has_stun
     self.stun_chance = stun_chance
     self.stun_duration = stun_duration
     self.has_fear = has_fear
     self.fear_chance = fear_chance
     self.fear_duration = fear_duration
     self.fear_damage_reduction = fear_damage_reduction
     self.has_lifesteal = has_lifesteal
     self.lifesteal_base = lifesteal_base
     self.lifesteal_per_con = lifesteal_per_con



 def __repr__(self):
     return self.name

# these are some weapons I added
START_DAGGER = Weapon(
 name="Kitchen Knife",
 weapon_type="Dagger",
 min_damage=3,
 max_damage=6,
 scaling_stats={"agility": 0.8, "strength": 0.4},
 allowed_classes=["Bandit"],
 dual_wield=True,
 light=True,
 description="not meant for battle but it can get the job done in the hands of a professional."
)
IRON_DAGGER = Weapon(
 name="Iron Dagger",
 weapon_type="Dagger",
 min_damage=5,
 max_damage=8,
 scaling_stats={"agility": 1.2, "strength": 0.8},
 allowed_classes=["Bandit"],
 allowed_enemies=["Assasin"],
 dual_wield=True,
 light=True,
 description="The customary weapon given to a freshly appointed Rouge."
)
GOLD_DAGGER = Weapon(
 name="Golden Dagger",
 weapon_type="Dagger",
 min_damage=7,
 max_damage=10,
 scaling_stats={"agility": 1.5, "strength": 1},
 allowed_classes=["Bandit"],
 dual_wield=True,
 light=True,
 description="A solid promotion deserves a solid gold upgrade."
)
DIAMOND_DAGGER = Weapon(
 name="Diamond Dagger",
 weapon_type="Dagger",
 min_damage=9,
 max_damage=11,
 scaling_stats={"agility": 1.7, "strength": 1.2},
 allowed_classes=["Bandit"],
 dual_wield=True,
 light=True,
 description="Reserved for the strongest of Rouges, an achievement only few can acheive."
)

LEGEND_DAGGER = Weapon(
 name="Eitr & Jörmungandr",
 weapon_type="Dagger",
 min_damage=9,
 max_damage=11,
 scaling_stats={"agility": 1.7, "strength": 1.2},
 allowed_classes=["Bandit"],
 dual_wield=True,
 light=True,
 has_poison=True, poison_chance=0.3, poison_duration=3, poison_damage_multiplier=0.05,
 description="Taken from the mouth of the Great Serpent, these dual-wielded daggers are known to secreet an uncurable poison that eats away at whoever is cut by it."
)

START_CLUB = Weapon(
 name="Wooden Club",
 weapon_type="Hammer",
 min_damage=10,
 max_damage=20,
 scaling_stats={"strength": 1.1},
 allowed_classes=["Tank"],
 heavy=True,
 description="a heavy wooden club, maybe too heavy, but the damage makes it worthwhile"
)
IRON_HAMMER = Weapon(
 name="Iron Hammer",
 weapon_type="Hammer",
 min_damage=13,
 max_damage=22,
 scaling_stats={"strength": 1.3},
 allowed_classes=["Tank"],
 allowed_enemies=["Ogre"],
 heavy=True,
 description="The customary weapon given to a freshly appointed Tank."
)
GOLD_HAMMER = Weapon(
 name="Golden Hammer",
 weapon_type="Hammer",
 min_damage=15,
 max_damage=25,
 scaling_stats={"strength": 1.5, "defense": 1},
 allowed_classes=["Tank"],
 heavy=True,
 description="A solid promotion deserves a solid gold upgrade."
)
DIAMOND_HAMMER = Weapon(
 name="Diamond Hammer",
 weapon_type="Hammer",
 min_damage=17,
 max_damage=27,
 scaling_stats={"strength": 1.7, "defense": 1.2},
 allowed_classes=["Tank"],
 heavy=True,
 description="Reserved for the strongest of Tanks, an achievement only few can acheive."
)
LEGEND_HAMMER = Weapon(
 name="Skull Crusher",
 weapon_type="Hammer",
 min_damage=17,
 max_damage=27,
 scaling_stats={"strength": 1.7, "defense": 1.2},
 allowed_classes=["Tank"],
 heavy=True,
 has_stun=True, stun_chance=0.5, stun_duration=1,
 description="Just as it name states, this hammer ways so much that hitting someone on the head breaks their skull. Why don't we test it out?"
)

START_WAND = Weapon(
 name="Cool stick",
 weapon_type="Wand",
 min_damage=6,
 max_damage=10,
 scaling_stats={"intelligence": 1.1},
 allowed_classes=["Mage"],
 description="A very cool stick, so cool in fact, that it can use magic through pure coolness"
)
MAGE_STAFF = Weapon(
 name="Iron Staff",
 weapon_type="Wand",
 min_damage=15,
 max_damage=10,
 scaling_stats={"intelligence": 1.3},
 allowed_classes=["Mage"],
 allowed_enemies=["Mage"],
 description="The customary weapon given to a freshly appointed Mage."
)
GOLD_STAFF = Weapon(
 name="Golden Staff",
 weapon_type="Wand",
 min_damage=20,
 max_damage=27,
 scaling_stats={"intelligence": 1.5, "constitution":1},
 allowed_classes=["Mage"],
 description="A solid promotion deserves a solid gold upgrade."
)
DIAMOND_STAFF = Weapon(
 name="Diamond Staff",
 weapon_type="Wand",
 min_damage=30,
 max_damage=40,
 scaling_stats={"intelligence": 1.7, "constitution":1.2},
 allowed_classes=["Mage"],
 description="Reserved for the strongest of Mages, an achievement only few can acheive."
)
LEGEND_WAND = Weapon(
 name="The Somber Grasp",
 weapon_type="Wand",
 min_damage=30,
 max_damage=40,
 scaling_stats={"intelligence": 1.7, "constitution":1.2},
 allowed_classes=["Mage"],
 has_fear=True, fear_chance=0.4, fear_duration=2, fear_damage_reduction=0.5,
 description="Gifted to man from the god of fear, Phobos, this sceptre invokes fear into the hearts of all enemies that come near it."
)

START_SWORD = Weapon(
 name="rusted sword",
 weapon_type="Sword",
 min_damage=6,
 max_damage=12,
 scaling_stats={"strength": 1.0},
 allowed_classes=["Warrior"],
 description="A worn but usable blade"
)
IRON_SWORD = Weapon(
 name="Iron Sword",
 weapon_type="Sword",
 min_damage=8,
 max_damage=14,
 scaling_stats={"strength": 1.1},
 allowed_classes=["Warrior"],
 allowed_enemies=["Knight"],
 description="The customary weapon given to a freshly appointed Warrior."
)
GOLD_SWORD = Weapon(
 name="Golden Sword",
 weapon_type="Sword",
 min_damage=10,
 max_damage=16,
 scaling_stats={"strength": 1.2, "agility": .8, "defense": .8, "constitution": .8, "intelligence": .8},
 allowed_classes=["Warrior"],
 description="A solid promotion deserves a solid gold upgrade."
)
DIAMOND_SWORD = Weapon(
 name="Diamond Sword",
 weapon_type="Sword",
 min_damage=12,
 max_damage=18,
 scaling_stats={"strength": 1.4, "agility": 1, "defense": 1, "constitution": 1, "intelligence": 1},
 allowed_classes=["Warrior"],
 description="Reserved for the strongest of Warriors, an achievement only few can acheive."
)
LEGEND_SWORD = Weapon(
 name="Skofnung",
 weapon_type="Sword",
 min_damage=12,
 max_damage=18,
 scaling_stats={"strength": 1.4, "agility": 1, "defense": 1, "constitution": 1, "intelligence": 1},
 allowed_classes=["Warrior"],
 has_lifesteal=True, lifesteal_base=1, lifesteal_per_con=1,
 description="Known for it's unquenchable thrist for blood, it will steal the life of it's enemies, and use their blood to heal it's owner's wounds."
)

LARRY_SWORD = Weapon(
    name="bone sword",
    weapon_type="sword",
    min_damage=2,
    max_damage=7,
    scaling_stats={"strength": 0.4, "agility": 0.6},
    allowed_enemies=["Skeleton"]
)
ZOMBIE_FIST = Weapon(
    name="Fist",
    weapon_type="fist",
    min_damage=1,
    max_damage=4,
    scaling_stats={"strength": 0, "agility": 0},
    allowed_enemies=["Zombie"]
)


GOBLIN_DAGGER = Weapon(
    name="goblin dagger",
    weapon_type="dagger",
    min_damage=1,
    max_damage=5,
    scaling_stats={"strength": 0.8},
    allowed_enemies=["Goblin"],
    dual_wield = True,
    light = True
)

DRAGON_CLAW=Weapon(
    name="Dragon Claw",
    weapon_type="Draconic",
    min_damage=50,
    max_damage=70,
    scaling_stats=None,
    allowed_classes=None,
    allowed_enemies=["Off Guard Dragon", "Dragon", "Tired Dragon"],
    description=("the dragon's claw, idk what you want me to say")
)


# Classes
MAGE_CLASS = CharacterClass(
 name="Mage",
 base_stats=Stats(
     strength=1,
     agility=2,
     defense=1,
     intelligence=6,
     constitution=2
 ),
 allowed_weapon_types=["Staff", "Wand"],
 description="Glass cannon spellcaster. High intelligence, low defense."
)




TANK = CharacterClass(
 name="Tank",
 base_stats=Stats(
     strength=5,
     agility=1,
     defense=6,
     intelligence=1,
     constitution=6
 ),
 allowed_weapon_types=["Mace", "Hammer","Club"],
 description="Slow, heavily armored bruiser with massive durability and strong but slow attacks."
)




WARRIOR = CharacterClass(
 name="Warrior",
 base_stats=Stats(
     strength=6,
     agility=3,
     defense=4,
     intelligence=1,
     constitution=4
 ),
 allowed_weapon_types=["Sword", "Axe"],
 description="Balanced melee fighter with strong offense and defense."
)




BANDIT = CharacterClass(
 name="Bandit",
 base_stats=Stats(
     strength=3,
     agility=6,
     defense=2,
     intelligence=1,
     constitution=3
 ),
 allowed_weapon_types=["Dagger"],
 description="Fast, agile attacker with high burst damage and average health."
)


#these are the enemies I added
Goblin=EnemyClass(
    name = "Goblin",
    base_stats=Stats(
        strength=3,
        agility=2,
        defense=1,
        intelligence=1,
        constitution=1,
    ),
    weapon = GOBLIN_DAGGER,
    exp = 2,
    min_hp = 10,
    description="A feisty green monster, who loves gold and causing mischief"
)
Skeleton=EnemyClass(
    name = "skeleton",
    base_stats=Stats(
        strength=2,
        agility=3,
        defense=1,
        intelligence=2,
        constitution=1,
    ),
    weapon = LARRY_SWORD,
    exp = 1,
    min_hp = 0,
    description="The dead corpse of a fallen knight, their soul still bound to combat"
)
Knight=EnemyClass(
    name = "Knight",
    base_stats=Stats(
        strength=4,
        agility=2,
        defense=3,
        intelligence=1,
        constitution=3,
    ),
    weapon = IRON_SWORD,
    exp = 10,
    min_hp = 30,
    description="A Knight in the empire's army, maybe a comrade once, but now they are the enemy",
    weapon_drop = IRON_SWORD
)
Assasin=EnemyClass(
    name = "Assasin",
    base_stats=Stats(
        strength=3,
        agility=4,
        defense=1,
        intelligence=2,
        constitution=1,
    ),
    weapon = IRON_DAGGER,
    exp = 10,
    min_hp = 20,
    description="An Assasin in the empire's army, you didn't like them when you were in the army, and you certainly don't like them now",
    weapon_drop = IRON_DAGGER
)
Ogre=EnemyClass(
    name = "Ogre",
    base_stats=Stats(
        strength=3,
        agility=1,
        defense=4,
        intelligence=1,
        constitution=4,
    ),
    weapon = IRON_HAMMER,
    exp = 20,
    min_hp = 15,
    description="A heavily armored beast with a massive iron hammer",
    weapon_drop = IRON_HAMMER
)
ROYAL_MAGE_ENEMY=EnemyClass(
    name = "Royal Mage",
    base_stats=Stats(
        strength=1,
        agility=3,
        defense=1,
        intelligence=5,
        constitution=1,
    ),
    weapon = MAGE_STAFF,
    exp = 20,
    min_hp = 25,
    description="A mage trained by the high mages of the empire, he's still learning but he knows enough spells to take you out",
    weapon_drop=MAGE_STAFF
)

Gold_Knight=EnemyClass(
    name = "Golden Knight",
    base_stats=Stats(
        strength=8,
        agility=8,
        defense=5,
        intelligence=4,
        constitution=5,
    ),
    weapon = GOLD_SWORD,
    exp = 20,
    min_hp = 35,
    description="Known for their stylish hairstyle, these looksmaxxers mog all their enemies with their crisp jawline.",
    weapon_drop = GOLD_SWORD
)

Gold_Tank=EnemyClass(
    name = "Golden Tank",
    base_stats=Stats(
        strength=12,
        agility=3,
        defense=8,
        intelligence=1,
        constitution=5,
    ),
    weapon = GOLD_HAMMER,
    exp = 20,
    min_hp = 35,
    description="These beasts were the 1st people in the 40-40-40 club; 40 kills, 40 cheeseburgers per meal, 40 pounds * 10",
    weapon_drop = GOLD_HAMMER
)
Gold_Mage=EnemyClass(
    name = "Golden Mage",
    base_stats=Stats(
        strength=6,
        agility=5,
        defense=2,
        intelligence=12,
        constitution=4,
    ),
    weapon = GOLD_STAFF,
    exp = 20,
    min_hp = 35,
    description="With an iq of 184, these mages have graduated prestigious schools in order to hone their intellect. ",
    weapon_drop = GOLD_STAFF
)
Gold_Bandit=EnemyClass(
    name = "Golden Bandit",
    base_stats=Stats(
        strength=8,
        agility=12,
        defense=3,
        intelligence=3,
        constitution=3,
    ),
    weapon = GOLD_DAGGER,
    exp = 20,
    min_hp = 35,
    description="'Stand and deliver!'",
    weapon_drop = GOLD_DAGGER
)

Diamond_Knight=EnemyClass(
    name = "General",
    base_stats=Stats(
        strength=10,
        agility=10,
        defense=7,
        intelligence=6,
        constitution=7,
    ),
    weapon = DIAMOND_SWORD,
    exp = 30,
    min_hp = 45,
    description="Commemorated for their battle acheivements, they have been promoted to commanded the leader's armies ",
    weapon_drop = DIAMOND_SWORD
)

Diamond_Tank=EnemyClass(
    name = "Nobleman",
    base_stats=Stats(
        strength=14,
        agility=5,
        defense=10,
        intelligence=3,
        constitution=7,
    ),
    weapon = DIAMOND_HAMMER,
    exp = 30,
    min_hp = 45,
    description="The biggest of big backs, known to consume 10,000+ calories per day",
    weapon_drop = DIAMOND_HAMMER
)
Diamond_Mage=EnemyClass(
    name = "Grand Vizier",
    base_stats=Stats(
        strength=8,
        agility=7,
        defense=4,
        intelligence=14,
        constitution=6,
    ),
    weapon = DIAMOND_STAFF,
    exp = 30,
    min_hp = 45,
    description="The advisor to the Caliphs of Arabia, these mages have the dark arts to wreak havoc against their enemies.",
    weapon_drop = DIAMOND_STAFF
)

Diamond_Bandit=EnemyClass(
    name = "Haberdasher",
    base_stats=Stats(
        strength=10,
        agility=14,
        defense=5,
        intelligence=5,
        constitution=5,
    ),
    weapon = DIAMOND_DAGGER,
    exp = 30,
    min_hp = 45,
    description="Clothes are their proffesion, selling anything from Gucci to your wallet.",
    weapon_drop = DIAMOND_DAGGER
)

Legend_Knight=EnemyClass(
    name = "Igris the Bloodred",
    base_stats=Stats(
        strength=20,
        agility=20,
        defense=15,
        intelligence=15,
        constitution=15,
    ),
    weapon = LEGEND_SWORD,
    exp = 500,
    min_hp = 350,
    description="Sung Jin-Woo may have passed away, but Igris still fights on for his new master",
    weapon_drop = LEGEND_SWORD
)

Legend_Tank=EnemyClass(
    name = "Clark Schutt",
    base_stats=Stats(
        strength=35,
        agility=20,
        defense=40,
        intelligence=10,
        constitution=30,
    ),
    weapon = LEGEND_HAMMER,
    exp = 500,
    min_hp = 350,
    description="With his refined eating skills, Clark has managed to gain 1000 pounds of pure muscle, crushing everyone in one blow",
    weapon_drop = LEGEND_HAMMER
)

Legend_Mage=EnemyClass(
    name = "The Necromancer",
    base_stats=Stats(
        strength=30,
        agility=30,
        defense=15,
        intelligence=40,
        constitution=20,
    ),
    weapon = LEGEND_WAND,
    exp = 500,
    min_hp = 350,
    description="After falling to the dark side, he honed his skills to be able to summon crazy amounts of skeletons for his army",
    weapon_drop = LEGEND_WAND
)

Legend_Bandit=EnemyClass(
    name = "Boss Bandit",
    base_stats=Stats(
        strength=40,
        agility=40,
        defense=20,
        intelligence=35,
        constitution=30,
    ),
    weapon = LEGEND_DAGGER,
    exp = 500,
    min_hp = 350,
    description="SUPERCELL NEEDS TO BE SLIMED OUT FOR CREATING THIS ABOMINATION",
    weapon_drop = LEGEND_DAGGER
)



Zombie=EnemyClass(
    name = "Zombie",
    base_stats=Stats(
        strength=4,
        agility=1,
        defense=3,
        intelligence=2,
        constitution=3,
    ),
    weapon = ZOMBIE_FIST,
    exp = 5,
    min_hp = 15,
    description="The rotten corpse of a local villager, their soul still bound to combat"
)
Off_Guard_Dragon=EnemyClass(
    name = "Off Guard Dragon",
    base_stats=Stats(
        strength=100,
        agility=25,
        defense=75,
        intelligence=50,
        constitution=75,
    ),
    weapon = DRAGON_CLAW,
    exp = 1000,
    min_hp = 100,
    description="A dragon, an off guard one, but still a dragon"
)
Dragon=EnemyClass(
    name = "Dragon",
    base_stats=Stats(
        strength=150,
        agility=40,
        defense=120,
        intelligence=80,
        constitution=120,
    ),
    weapon = DRAGON_CLAW,
    exp = 1000,
    min_hp = 100,
    description="You managed to beat him once, but I wouldn't get to confident, I think you just made him mad."
)
Tired_Dragon=EnemyClass(
    name = "Dragon",
    base_stats=Stats(
        strength=75,
        agility=15,
        defense=50,
        intelligence=15,
        constitution=50,
    ),
    weapon = DRAGON_CLAW,
    exp = 1000,
    min_hp = 100,
    description="Well, you have managed to live this long, just a bit more and you might just win, or you might explode"
)
#here is some armor I added
START_CHESTPLATE=armor(
    name = "Prison Jumpsuit Shirt",
    Armor_type = "chestplate",
    Protection=2,
    defense_requirement=1,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="A prison jumpsuit, might save you from a few punches but if the enemy has a weapon your dead",
    armor_light=True
)

START_PANTS=armor(
    name = "Prison Jumpsuit Pants",
    Armor_type = "pants",
    Protection=2,
    defense_requirement=1,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="A prison jumpsuit, might save you from a few punches but if the enemy has a weapon your dead",
    armor_light=True
)

START_BOOTS=armor(
    name = "Boots",
    Armor_type = "boots",
    Protection=2,
    defense_requirement=1,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="The empire might have taken everything else from you, but at least you still have your shoes!",
    armor_light=True
)

IRON_HELMET=armor(
    name = "Iron Helmet",
    Armor_type = "helmet",
    Protection=5,
    defense_requirement=8,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="The standard armor given to new recruits",
    armor_heavy=True
)
IRON_CHESTPLATE=armor(
    name = "Iron Chestplate",
    Armor_type = "chestplate",
    Protection=5,
    defense_requirement=8,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="The standard armor given to new recruits",
    armor_heavy=True
)

IRON_PANTS=armor(
    name = "Iron Pants",
    Armor_type = "pants",
    Protection=5,
    defense_requirement=8,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="The standard armor given to new recruits",
    armor_heavy=True
)

IRON_BOOTS=armor(
    name = "Iron Boots",
    Armor_type = "boots",
    Protection=5,
    defense_requirement=8,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="The standard armor given to new recruits",
    armor_heavy=True
)

GOLD_HELMET=armor(
    name = "Golden Helmet",
    Armor_type = "helmet",
    Protection=8,
    defense_requirement=12,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="You can finally loot the treasure bastion",
    armor_heavy=True
)
GOLD_CHESTPLATE=armor(
    name = "Golden Helmet",
    Armor_type = "chestplate",
    Protection=8,
    defense_requirement=12,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="You can finally loot the treasure bastion",
    armor_heavy=True
)

GOLD_PANTS=armor(
    name = "Golden Pants",
    Armor_type = "pants",
    Protection=8,
    defense_requirement=12,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="You can finally loot the treasure bastion",
    armor_heavy=True
)

GOLD_BOOTS=armor(
    name = "Golden Boots",
    Armor_type = "boots",
    Protection=8,
    defense_requirement=12,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="You can finally loot the treasure bastion",
    armor_heavy=True
)
DIAMOND_HELMET=armor(
    name = "Diamond Helmet",
    Armor_type = "helmet",
    Protection=12,
    defense_requirement=16,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="They told me bust out my AP, PERFECT TIMING!",
    armor_heavy=True
)
DIAMOND_CHESTPLATE=armor(
    name = "Diamond Chestplate",
    Armor_type = "chestplate",
    Protection=12,
    defense_requirement=16,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="They told me bust out my AP, PERFECT TIMING!",
    armor_heavy=True
)
DIAMOND_PANTS=armor(
    name = "Diamond Pants",
    Armor_type = "pants",
    Protection=12,
    defense_requirement=16,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="They told me bust out my AP, PERFECT TIMING!",
    armor_heavy=True
)
DIAMOND_BOOTS=armor(
    name = "Diamond Boots",
    Armor_type = "boots",
    Protection=12,
    defense_requirement=16,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="They told me bust out my AP, PERFECT TIMING!",
    armor_heavy=True
)
LEGEND_HELMET=armor(
    name = "Skull Protector",
    Armor_type = "helmet",
    Protection=20,
    defense_requirement=20,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="Now you can crush skulls while protecting yours",
    armor_heavy=True
)
LEGEND_CHESTPLATE=armor(
    name = "Heaven's Release",
    Armor_type = "chestplate",
    Protection=20,
    defense_requirement=20,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="When Lucifer fell, the archangels crafted this vest to protect humans from the call of Satan",
    armor_heavy=True
)
LEGEND_PANTS=armor(
    name = "Hypaepa's Vraka",
    Armor_type = "pants",
    Protection=20,
    defense_requirement=20,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="Made from spindle fibers spun from Arachne's thread, these pants are harder than steel",
    armor_heavy=True
)
LEGEND_BOOTS=armor(
    name = "Draugar's Skor",
    Armor_type = "boots",
    Protection=20,
    defense_requirement=20,
    allowed_classes=["Warrior", "Bandit", "Tank", "Mage"],
    description="Forged with the blood of a deceased Vampire, these boots grant protection against damage",
    armor_heavy=True
)
#items
SMALL_HEALTH_POT = Items(
    name="Small Health Potion",
    Is_Health_Pot=True,
    Health_healed=20,
    description="Heals 20 HP"
)

MED_HEALTH_POT = Items(
    name="Medium Health Potion",
    Is_Health_Pot=True,
    Health_healed=60,
    description="Heals 60 HP"
)

LARGE_HEALTH_POT = Items(
    name="Large Health Potion",
    Is_Health_Pot=True,
    Health_healed=120,
    description="Heals 120 HP"
)

MAX_HEALTH_POT = Items(
    name="Maximum Health Potion",
    Is_Health_Pot=True,
    Health_healed=None, # Changed to None to be dynamically calculated at use time
    description="Heals All HP"
)

SMALL_STRENGTH_POT = Items(
    name="Small Strength Potion",
    Is_Strength_pot=True,
    Stat_Buff=10,
    description="Buffs Strength by 10 points"
)

MED_STRENGTH_POT = Items(
    name="Medium Strength Potion",
    Is_Strength_pot=True,
    Stat_Buff=25,
    description="Buffs Strength by 25 points"
)

LARGE_STRENGTH_POT = Items(
    name="Large Strength Potion",
    Is_Strength_pot=True,
    Stat_Buff=45,
    description="Buffs Strength by 45 points"
)

XL_STRENGTH_POT = Items(
    name="Extra Large Strength Potion",
    Is_Strength_pot=True,
    Stat_Buff=60,
    description="Buffs Strength by 60 points"
)

SMALL_SPEED_POT = Items(
    name="Small Speed Potion",
    Is_speed_pot=True,
    Stat_Buff=10,
    description="Buffs Speed by 10 points"
)

MED_SPEED_POT = Items(
    name="Medium Speed Potion",
    Is_speed_pot=True,
    Stat_Buff=25,
    description="Buffs Speed by 25 points"
)

LARGE_SPEED_POT = Items(
    name="Large Speed Potion",
    Is_speed_pot=True,
    Stat_Buff=45,
    description="Buffs Speed by 45 points"
)

XL_SPEED_POT = Items(
    name="Extra Large Speed Potion",
    Is_speed_pot=True,
    Stat_Buff=60,
    description="Buffs Speed by 60 points"
)

SMALL_DEFENSE_POT = Items(
    name="Small Defense Potion",
    Is_Defense_pot=True,
    Stat_Buff=10,
    description="Buffs Defense by 10 points"
)

MED_DEFENSE_POT = Items(
    name="Medium Defense Potion",
    Is_Defense_pot=True,
    Stat_Buff=25,
    description="Buffs Defense by 25 points"
)

LARGE_DEFENSE_POT = Items(
    name="Large Defense Potion",
    Is_Defense_pot=True,
    Stat_Buff=45,
    description="Buffs Defense by 45 points"
)

XL_DEFENSE_POT = Items(
    name="Extra Large Defense Potion",
    Is_Defense_pot=True,
    Stat_Buff=60,
    description="Buffs Defense by 60 points"
)

SMALL_INTELLIGENCE_POT = Items(
    name="Small Intelligence Potion",
    Is_Intelligence_pot=True,
    Stat_Buff=10,
    description="Buffs Intelligence by 10 points"
)

MED_INTELLIGENCE_POT = Items(
    name="Medium Intelligence Potion",
    Is_Intelligence_pot=True,
    Stat_Buff=25,
    description="Buffs Intelligence by 25 points"
)

LARGE_INTELLIGENCE_POT = Items(
    name="Large Intelligence Potion",
    Is_Intelligence_pot=True,
    Stat_Buff=45,
    description="Buffs Intelligence by 45 points"
)

XL_INTELLIGENCE_POT = Items(
    name="Extra Large Intelligence Potion",
    Is_Intelligence_pot=True,
    Stat_Buff=60,
    description="Buffs Intelligence by 60 points"
)


CLASS_BOOK=Items(
    name="Ancient tome",
    Is_Book=True,
    description="An ancient torn and withered book in the corner of your cell"
)

# Rooms - Define all rooms first with None connections
START_CELL = Room(
    name="Start cell",
    items=[CLASS_BOOK],
    weapons=[],
    enemies=[],
    armor=[],
    description="the cell you spawn in"
)
Hlw1 = Room(
  name="Hallway 1",
  items=[],
  weapons=[],
  enemies=[Skeleton],
  armor=[],
  description="Hallway outside of cell"
)
Hlw2 = Room(
    name="Hallway 2",
    items=[],
    weapons=[],
    enemies=[],
    armor=[],
    description="Another hallway"
)
Hlw3 = Room(
    name="Hallway 3",
    items=[],
    weapons=[],
    enemies=[],
    armor=[],
    description="Another hallway"
)
Hlw4 = Room(
    name="Hallway 4",
    items=[],
    weapons=[],
    enemies=[Ogre],
    armor=[],
    description="Another hallway"
)
Control_room = Room(
    name="Mission Control Room",
    items=[SMALL_HEALTH_POT],
    weapons=[],
    enemies=[Knight, Knight, Knight,ROYAL_MAGE_ENEMY],
    armor=[GOLD_HELMET,GOLD_CHESTPLATE,GOLD_PANTS,GOLD_BOOTS],
    description="The HQ of the entire dungeon"
    )
Hlw5 = Room(
    name="Hallway 5",
    items=[],
    weapons=[],
    enemies=[Goblin],
    armor=[],
    description="Another hallway"
    )
Hlw6 = Room(
    name="Hallway 6",
    items=[],
    weapons=[],
    enemies=[],
    armor=[],
    description="Another hallway"
    )
Hlw7 = Room(
    name="Hallway 7",
    items=[],
    weapons=[],
    enemies=[],
    armor=[],
    description="Another hallway"
    )
Hlw9 = Room(
    name="Hallway 9",
    items=[],
    weapons=[],
    enemies=[],
    armor=[],
    description="Another hallway"
)
Hlw11 = Room(
    name="Hallway 11",
    items=[],
    weapons=[],
    enemies=[Goblin],
    armor=[],
    description="Another hallway"
    )
Hlw13 = Room(
    name="Hallway 13",
    items=[],
    weapons=[],
    enemies=[Knight],
    armor=[],
    description="Another hallway"
    )
Hlw10 = Room(
    name="Hallway 10",
    items=[],
    weapons=[],
    enemies=[Goblin],
    armor=[],
    description="Another hallway"
    )
Hlw8 = Room(
    name="Hallway 8",
    items=[],
    weapons=[],
    enemies=[],
    armor=[],
    description="Another hallway"
    )
Hlw12 = Room(
    name="Hallway 12",
    items=[],
    weapons=[],
    enemies=[],
    armor=[],
    description="Another hallway"
)
Hlw14 = Room(name="Hallway 14",
    items=[],
    weapons=[],
    enemies=[Skeleton],
    armor=[],
    description="Another hallway"
)
Pc1 = Room(
    name="Prison Cell 1",
    items=[SMALL_HEALTH_POT],
    weapons=[],
    enemies=[Skeleton],
    armor=[],
    description="A empty cell... Or is it?"
)
Pc2 = Room(
    name="Prison Cell 2",
    items=[],
    weapons=[],
    enemies=[Skeleton],
    armor=[IRON_PANTS],
    description="A cold, despairing cell"
)
Pc3 = Room(
    name="Prison Cell 3",
    items=[],
    weapons=[],
    enemies=[Skeleton],
    armor=[IRON_PANTS],
    description="A cold, despairing cell"
)
Pc4 = Room(
    name="Prison Cell 4",
    items=[],
    weapons=[IRON_HAMMER],
    enemies=[Knight],
    armor=[IRON_BOOTS],
    description="A cold cell"
)
Pc5 = Room(
    name="Prison Cell 5",
    items=[],
    weapons=[IRON_DAGGER],
    enemies=[Assasin],
    armor=[IRON_BOOTS, IRON_CHESTPLATE, SMALL_HEALTH_POT],
    description="A cold cell"
)
Pc6 = Room(
    name="Prison Cell 6",
    items=[SMALL_HEALTH_POT],
    weapons=[IRON_DAGGER],
    enemies=[Goblin],
    armor=[IRON_BOOTS, IRON_CHESTPLATE],
    description="A cold cell"
)
Pc7 = Room(
    name="Prison Cell 7",
    items=[SMALL_HEALTH_POT],
    weapons=[MAGE_STAFF],
    enemies=[Assasin],
    armor=[IRON_BOOTS, IRON_HELMET],
    description="A cold cell"
)
Staircase1 = Room(
    name="Staircase Level 1",
    items=[],
    weapons=[],
    enemies=[Ogre],
    armor=[],
    description="A staircase leading up"
)
Staircase2=Room(
    name="Staircase",
    items=[LARGE_HEALTH_POT],
    weapons=[DIAMOND_SWORD],
    enemies=[Diamond_Knight],
    armor=[DIAMOND_PANTS]
)
Armory=Room(
    name="Armory",
    items=[LARGE_HEALTH_POT, LARGE_STRENGTH_POT, LARGE_DEFENSE_POT],
    weapons=[DIAMOND_SWORD, DIAMOND_HAMMER, DIAMOND_STAFF],
    enemies=[Gold_Knight, Gold_Mage],
    armor=[DIAMOND_HELMET, DIAMOND_CHESTPLATE, DIAMOND_BOOTS],
    description="the armory with lots of stuff"
)
Pc15=Room(
    name="Cell",
    items=[SMALL_HEALTH_POT, MED_DEFENSE_POT],
    weapons=[IRON_DAGGER],
    enemies=[Goblin],
    armor=[IRON_HELMET],
    description="a cold cell"
)
Hlw23=Room(
    name="Hallway",
    items=[MED_SPEED_POT, SMALL_HEALTH_POT],
    weapons=[IRON_SWORD],
    enemies=[Skeleton, Goblin],
    armor=[IRON_HELMET],
    description="a hallway"
)
Hlw21=Room(
    name="Hallway",
    items=[SMALL_SPEED_POT],
    weapons=[IRON_SWORD],
    enemies=[Skeleton],
    armor=[IRON_BOOTS],
    description="a hallway"
)
Pc13=Room(
    name="Cell",
    items=[SMALL_STRENGTH_POT],
    weapons=[IRON_HAMMER],
    enemies=[Goblin],
    armor=[IRON_CHESTPLATE],
    description="a cold cell"
)
Pc11=Room(
    name="Cell",
    items=[SMALL_DEFENSE_POT],
    weapons=[IRON_SWORD],
    enemies=[Skeleton],
    armor=[IRON_BOOTS],
    description="a cold cell"
)
Hlw19=Room(
    name="Hallway",
    items=[MED_HEALTH_POT],
    weapons=[IRON_HAMMER],
    enemies=[Zombie],
    armor=[IRON_PANTS],
    description="a hallway"
)
Hlw17=Room(
    name="Hallway",
    items=[MED_HEALTH_POT],
    weapons=[IRON_HAMMER],
    enemies=[Ogre],
    armor=[IRON_CHESTPLATE],
    description="a hallway"
)
Pc9=Room(
    name="Cell",
    items=[SMALL_HEALTH_POT],
    weapons=[IRON_DAGGER],
    enemies=[Goblin],
    armor=[IRON_HELMET],
    description="a cold cell"
)
Hlw25=Room(
    name="Hallway",
    items=[MED_STRENGTH_POT],
    weapons=[GOLD_SWORD],
    enemies=[Knight],
    armor=[GOLD_HELMET],
    description="a hallway"
)
Hlw15=Room(
    name="Hallway",
    items=[MED_SPEED_POT],
    weapons=[GOLD_STAFF],
    enemies=[Gold_Mage],
    armor=[GOLD_HELMET],
    description="a hallway"
)
Exit=Room(
    name="Exit",
    items=[MAX_HEALTH_POT, LARGE_SPEED_POT],
    weapons=[DIAMOND_SWORD],
    enemies=[Gold_Knight],
    armor=[DIAMOND_CHESTPLATE],
    description="Yay the exit!"
)
Hlw28=Room(
    name="Hallway",
    items=[SMALL_SPEED_POT],
    weapons=[IRON_SWORD],
    enemies=[Skeleton],
    armor=[IRON_BOOTS],
    description="a hallway"
)
Hlw27=Room(
    name="Hallway",
    items=[MED_HEALTH_POT],
    weapons=[IRON_DAGGER],
    enemies=[Skeleton],
    armor=[IRON_BOOTS],
    description="a hallway"
)
Hlw26=Room(
    name="Hallway",
    items=[MED_SPEED_POT],
    weapons=[IRON_HAMMER],
    enemies=[Zombie],
    armor=[IRON_PANTS],
    description="a hallway"
)
Hlw16=Room(
    name="Hallway",
    items=[MED_STRENGTH_POT],
    weapons=[GOLD_SWORD],
    enemies=[Gold_Tank],
    armor=[GOLD_CHESTPLATE],
    description="a hallway"
)
Pc14=Room(
    name="Cell",
    items=[MED_HEALTH_POT],
    weapons=[GOLD_DAGGER],
    enemies=[Zombie],
    armor=[IRON_PANTS],
    description="a cold cell"
)
Hlw24=Room(
    name="Hallway",
    items=[MED_HEALTH_POT],
    weapons=[GOLD_SWORD],
    enemies=[Knight],
    armor=[GOLD_BOOTS],
    description="a hallway"
)
Hlw22=Room(
    name="Hallway",
    items=[MED_INTELLIGENCE_POT],
    weapons=[MAGE_STAFF],
    enemies=[ROYAL_MAGE_ENEMY],
    armor=[GOLD_HELMET],
    description="a hallway"
)
Hlw20=Room(
    name="Hallway",
    items=[MED_SPEED_POT],
    weapons=[GOLD_DAGGER],
    enemies=[Assasin],
    armor=[GOLD_PANTS],
    description="a hallway"
)
Hlw18=Room(
    name="Hallway",
    items=[MED_HEALTH_POT],
    weapons=[IRON_SWORD],
    enemies=[Zombie],
    armor=[IRON_BOOTS],
    description="a hallway"
)
Pc8=Room(
    name="Cell",
    items=[MED_DEFENSE_POT],
    weapons=[IRON_SWORD],
    enemies=[Skeleton],
    armor=[IRON_BOOTS],
    description="a cold cell"
)
Pc12=Room(
    name="Cell",
    items=[SMALL_DEFENSE_POT],
    weapons=[IRON_DAGGER],
    enemies=[Goblin],
    armor=[IRON_HELMET],
    description="a cold cell"
)
Pc10=Room(
    name="Cell",
    items=[SMALL_HEALTH_POT],
    weapons=[IRON_DAGGER],
    enemies=[Goblin],
    armor=[IRON_HELMET],
    description="a cold cell"
)

Dragon_Room=Room(
    name="Dragon's chamber",
    items=[],
    weapons=[],
    enemies=[],
    armor=[],
    description="A massive cavern filled to the brim with gold and jewels",
    is_boss_room=True
)
Mb1 = Room(name="Hall of  Boss Bandit",
    items=[MAX_HEALTH_POT],
    weapons=[LEGEND_DAGGER],
    enemies=[Legend_Bandit],
    armor=[LEGEND_PANTS],
    description="Home of Boss Bandit "
)
Mb2 = Room(
    name="Hall of Clark Schutt",
    items=[MAX_HEALTH_POT],
    weapons=[LEGEND_HAMMER],
    enemies=[Legend_Tank],
    armor=[LEGEND_HELMET],
    description="Home of Clark Schutt"
    )
Mb3 = Room(
    name="Hall of The Necromancer",
    items=[MAX_HEALTH_POT],
    weapons=[LEGEND_WAND],
    enemies=[Legend_Mage],
    armor=[IRON_HELMET, IRON_BOOTS, IRON_PANTS],
    description="Home of The Necromancer"
)
Mb4 = Room(
    name="Hall of Igris the Bloodred",
    items=[MAX_HEALTH_POT],
    weapons=[LEGEND_SWORD],
    enemies=[Legend_Knight],
    armor=[LEGEND_BOOTS],
    description="Home of Igris the Bloodred"
)
START_CELL.North_connection = Hlw1.name

Hlw1.North_connection = None
Hlw1.East_connection = Hlw2.name
Hlw1.South_connection = START_CELL.name
Hlw1.West_connection = Hlw3.name

Hlw2.North_connection = Hlw4.name
Hlw2.East_connection = None
Hlw2.South_connection = None
Hlw2.West_connection = Hlw1.name

Hlw3.North_connection = None
Hlw3.East_connection = Hlw1.name
Hlw3.South_connection = None
Hlw3.West_connection = None

Hlw4.North_connection = Control_room.name
Hlw4.East_connection = None
Hlw4.South_connection = Hlw2.name
Hlw4.West_connection = None

Control_room.North_connection = None
Control_room.East_connection = None
Control_room.South_connection = Hlw4.name
Control_room.West_connection = None

Hlw5.North_connection = Pc1.name
Hlw5.East_connection = Hlw3.name
Hlw5.South_connection = Hlw7.name
Hlw5.West_connection = None

Hlw6.North_connection = None
Hlw6.East_connection = None #
Hlw6.South_connection = Hlw8.name
Hlw6.West_connection = Hlw2.name

Hlw7.North_connection = Hlw5.name
Hlw7.East_connection = None
Hlw7.South_connection = Hlw9.name
Hlw7.West_connection = Pc3.name

Hlw9.North_connection = Hlw7.name
Hlw9.East_connection = None
Hlw9.South_connection = Hlw11.name
Hlw9.West_connection = None

Hlw11.North_connection = Hlw9.name
Hlw11.East_connection = Hlw13.name
Hlw11.South_connection = Pc7.name
Hlw11.West_connection = Pc5.name

Hlw13.North_connection = None
Hlw13.East_connection = Staircase1.name
Hlw13.South_connection = None
Hlw13.West_connection = Hlw11.name

Hlw10.North_connection = Hlw8.name
Hlw10.East_connection = Pc4.name
Hlw10.South_connection = Hlw12.name
Hlw10.West_connection = None

Hlw8.North_connection = Hlw6.name
Hlw8.East_connection = None
Hlw8.South_connection = Hlw10.name
Hlw8.West_connection = None

Hlw12.North_connection = Hlw10.name
Hlw12.East_connection = None
Hlw12.South_connection = Pc6.name
Hlw12.West_connection = Hlw14.name

Hlw14.North_connection = None
Hlw14.East_connection = Hlw12.name
Hlw14.South_connection = None
Hlw14.West_connection = Staircase1.name

Pc1.North_connection = None
Pc1.East_connection = None
Pc1.South_connection = Hlw5.name
Pc1.West_connection = None

Pc2.North_connection = None
Pc2.East_connection = None
Pc2.South_connection = None
Pc2.West_connection = Hlw6.name

Pc3.North_connection = None
Pc3.East_connection = Hlw7.name
Pc3.South_connection = None
Pc3.West_connection = None

Pc4.North_connection = None
Pc4.East_connection = None
Pc4.South_connection = None
Pc4.West_connection = Hlw10.name

Pc5.North_connection = None
Pc5.East_connection = Hlw11.name
Pc5.South_connection = None
Pc5.West_connection = None

Pc6.North_connection = Hlw12.name
Pc6.East_connection = None
Pc6.South_connection = None
Pc6.West_connection = None

Pc7.North_connection = Hlw11.name
Pc7.East_connection = None
Pc7.South_connection = None
Pc7.West_connection = None

Staircase1.North_connection = Staircase2.name
Staircase1.East_connection = Hlw14.name
Staircase1.South_connection = None
Staircase1.West_connection = Hlw13.name

Staircase2.North_connection = Armory.name
Staircase2.East_connection = Hlw16.name
Staircase2.South_connection = Staircase1.name
Staircase2.West_connection = Hlw15.name

Armory.North_connection = None
Armory.East_connection = None
Armory.South_connection = Staircase2.name
Armory.West_connection = None

Pc15.North_connection = None
Pc15.East_connection = None
Pc15.South_connection = Hlw23.name
Pc15.West_connection = None

Hlw23.North_connection = Pc15.name
Hlw23.East_connection = None
Hlw23.South_connection = Hlw21.name
Hlw23.West_connection = None

Hlw21.North_connection = Hlw23.name
Hlw21.East_connection = None
Hlw21.South_connection = Hlw19.name
Hlw21.West_connection = Pc13.name

Pc13.North_connection = None
Pc13.East_connection = Hlw21.name
Pc13.South_connection = Pc11.name
Pc13.West_connection = None

Pc11.North_connection = Pc13.name
Pc11.East_connection = Hlw17.name
Pc11.South_connection = None
Pc11.West_connection = None

Hlw19.North_connection = Hlw21.name
Hlw19.East_connection = None
Hlw19.South_connection = Hlw17.name
Hlw19.West_connection = None

Hlw17.North_connection = Hlw19.name
Hlw17.East_connection = Hlw15.name
Hlw17.South_connection = Pc9.name
Hlw17.West_connection = Pc11.name

Pc9.North_connection = Hlw17.name
Pc9.East_connection = None
Pc9.South_connection = None
Pc9.West_connection = None

Hlw25.North_connection = None
Hlw25.East_connection = Hlw27.name
Hlw25.South_connection = Hlw15.name
Hlw25.West_connection = Hlw23.name

Hlw15.North_connection = Hlw25.name
Hlw15.East_connection = Staircase2.name
Hlw15.South_connection = None
Hlw15.West_connection = Hlw17.name

Exit.North_connection = None
Exit.East_connection = None
Exit.South_connection = Hlw28.name
Exit.West_connection = None

Hlw28.North_connection = Exit.name
Hlw28.East_connection = None
Hlw28.South_connection = Hlw27.name
Hlw28.West_connection = None

Hlw27.North_connection = Hlw28.name
Hlw27.East_connection = Hlw26.name
Hlw27.South_connection = Armory.name
Hlw27.West_connection = Hlw25.name

Hlw26.North_connection = None
Hlw26.East_connection = Hlw24.name
Hlw26.South_connection = Hlw16.name
Hlw26.West_connection = Hlw27.name

Hlw16.North_connection = Hlw26.name
Hlw16.East_connection = Hlw18.name
Hlw16.South_connection = None
Hlw16.West_connection = Staircase2.name

Pc14.North_connection = None
Pc14.East_connection = None
Pc14.South_connection = Hlw24.name
Pc14.West_connection = None

Hlw24.North_connection = Pc14.name
Hlw24.East_connection = Pc12.name
Hlw24.South_connection = Hlw22.name
Hlw24.West_connection = Hlw26.name

Hlw22.North_connection = Hlw24.name
Hlw22.East_connection = None
Hlw22.South_connection = Hlw20.name
Hlw22.West_connection = None

Hlw20.North_connection = Hlw22.name
Hlw20.East_connection = Pc10.name
Hlw20.South_connection = Hlw18.name
Hlw20.West_connection = None

Hlw18.North_connection = Hlw20.name
Hlw18.East_connection = None
Hlw18.South_connection = Pc8.name
Hlw18.West_connection = Hlw16.name

Pc8.North_connection = Hlw18.name
Pc8.East_connection = None
Pc8.South_connection = None
Pc8.West_connection = None

Pc12.North_connection = None
Pc12.East_connection = None
Pc12.South_connection = Pc10.name
Pc12.West_connection = Hlw24.name

Pc10.North_connection = Pc12.name
Pc10.East_connection = None
Pc10.South_connection = None
Pc10.West_connection = Hlw20.name

Mb1.North_connection = Pc13.name
Mb1.East_connection = Hlw19.name
Mb1.South_connection = Pc11.name
Mb1.West_connection = None

Mb3.North_connection = Pc3.name
Mb3.East_connection = Hlw9.name
Mb3.South_connection = Pc5.name
Mb3.West_connection = None

Mb2.North_connection = Pc12.name
Mb2.East_connection = Hlw22.name
Mb2.South_connection = Pc10.name
Mb2.West_connection = None

Mb4.North_connection = Pc2.name
Mb4.East_connection = Hlw8.name
Mb4.South_connection = Pc4.name
Mb4.West_connection = None

# Define a dictionary of all rooms after all rooms have been instantiated
all_rooms = {
    START_CELL.name: START_CELL,
    Hlw1.name: Hlw1,
    Hlw2.name: Hlw2,
    Hlw3.name: Hlw3,
    Hlw4.name: Hlw4,
    Control_room.name: Control_room,
    Hlw5.name: Hlw5,
    Hlw6.name: Hlw6,
    Hlw7.name: Hlw7,
    Hlw9.name: Hlw9,
    Hlw11.name: Hlw11,
    Hlw13.name: Hlw13,
    Hlw10.name: Hlw10,
    Hlw8.name: Hlw8,
    Hlw12.name: Hlw12,
    Hlw14.name: Hlw14,
    Pc1.name: Pc1,
    Pc2.name: Pc2,
    Pc3.name: Pc3,
    Pc4.name: Pc4,
    Pc5.name: Pc5,
    Pc6.name: Pc6,
    Pc7.name: Pc7,
    Staircase1.name: Staircase1,
    Staircase2.name: Staircase2,
    Armory.name: Armory,
    Pc15.name: Pc15,
    Hlw23.name: Hlw23,
    Hlw21.name: Hlw21,
    Pc13.name: Pc13,
    Pc11.name: Pc11,
    Hlw19.name: Hlw19,
    Hlw17.name: Hlw17,
    Pc9.name: Pc9,
    Hlw25.name: Hlw25,
    Hlw15.name: Hlw15,
    Exit.name: Exit,
    Hlw28.name: Hlw28,
    Hlw27.name: Hlw27,
    Hlw26.name: Hlw26,
    Hlw16.name: Hlw16,
    Pc14.name: Pc14,
    Hlw24.name: Hlw24,
    Hlw22.name: Hlw22,
    Hlw20.name: Hlw20,
    Hlw18.name: Hlw18,
    Pc8.name: Pc8,
    Pc12.name: Pc12,
    Pc10.name: Pc10,
    Dragon_Room.name: Dragon_Room,
    Mb1.name: Mb1,
    Mb2.name: Mb2,
    Mb3.name: Mb3,
    Mb4.name: Mb4
}

#this is for the player
class Player:
    def __init__(self, name, character_class_str):
        self.name = name
        self.level = 1
        self.gold = 0
        self.xp = 0
        self.inventory = []
        self.current_room = None

        class_mapping = {
            "warrior": (WARRIOR, START_SWORD),
            "bandit": (BANDIT, START_DAGGER),
            "tank": (TANK, START_CLUB),
            "mage": (MAGE_CLASS, START_WAND),
        }

        char_class_obj, starting_weapon = class_mapping.get(character_class_str.lower())

        if not char_class_obj:
            raise ValueError(f"Unknown character class: {character_class_str}")

        if starting_weapon.weapon_type not in char_class_obj.allowed_weapon_types:
            raise ValueError(f"{char_class_obj.name} cannot use {starting_weapon.name}")

        self.character_class = char_class_obj
        self.stats = char_class_obj.base_stats
        self.weapon = starting_weapon
        self.second_weapon = None # Added for dual-wielding
        self.hp = self.stats.max_hp()
        self.max_hp = self.stats.max_hp()

    def get_attack(self):
        if self.character_class.name == "Warrior":
            return self.stats.strength
        elif self.character_class.name == "Bandit":
            return self.stats.agility
        elif self.character_class.name == "Tank":
            return self.stats.strength
        elif self.character_class.name == "Mage":
            return self.stats.intelligence
        return 1

    def get_defense(self):
        return self.stats.defense

    def __repr__(self):
        return self.name

#this is the combat system
def calculate_damage(attacker_obj, weapon, defender_obj):

    base_damage = random.randint(weapon.min_damage, weapon.max_damage)

    scaling_bonus = 0
    if weapon.scaling_stats:
        attacker_stats_to_use = attacker_obj.stats if isinstance(attacker_obj, Player) else attacker_obj.base_stats
        for stat, scale in weapon.scaling_stats.items():
            scaling_bonus += getattr(attacker_stats_to_use, stat) * scale

    damage = base_damage + scaling_bonus

    defender_stats_to_use = defender_obj.stats if isinstance(defender_obj, Player) else defender_obj.base_stats
    final_damage = max(1, int(damage) - int(defender_stats_to_use.defense * 0.5)) # Ensure at least 1 damage

    # Apply fear reduction if enemy is feared
    if isinstance(defender_obj, EnemyClass) and defender_obj.is_feared and defender_obj.fear_remaining_turns > 0:
        final_damage = int(final_damage * (1 - defender_obj.feared_damage_reduction))

    return final_damage
def use_potion(player):
    # Filter for all types of potions
    available_potions = [item for item in player.inventory if any([item.Is_Health_Pot, item.Is_Strength_pot, item.Is_Defense_pot, item.Is_Intelligence_pot, item.Is_speed_pot])]

    if not available_potions:
        slow_print("You have no potions!")
        time.sleep(1)
        return False

    slow_print("\nWhich potion would you like to use?")
    for i, potion in enumerate(available_potions):
        slow_print(f"[{i+1}] {potion.name}: {potion.description}")
    slow_print("[0] Cancel")

    while True:
        try:
            choice = int(input("> "))
            if choice == 0:
                slow_print("Potion usage cancelled.")
                time.sleep(1)
                return False
            if 1 <= choice <= len(available_potions):
                chosen_potion = available_potions[choice - 1]
                break
            else:
                slow_print("Invalid choice. Please enter a number from the list.")
        except ValueError:
            slow_print("Invalid input. Please enter a number.")

    # Apply potion effect
    if chosen_potion.Is_Health_Pot:
        if chosen_potion.Health_healed is None:
            heal = player.max_hp - player.hp
        else:
            heal = chosen_potion.Health_healed
        player.hp = min(player.max_hp, player.hp + heal)
        slow_print(f"You used {chosen_potion.name} and healed {heal} HP!")
    elif chosen_potion.Is_Strength_pot:
        player.stats.strength += chosen_potion.Stat_Buff
        slow_print(f"You used {chosen_potion.name} and gained {chosen_potion.Stat_Buff} Strength!")
    elif chosen_potion.Is_Defense_pot:
        player.stats.defense += chosen_potion.Stat_Buff
        slow_print(f"You used {chosen_potion.name} and gained {chosen_potion.Stat_Buff} Defense!")
    elif chosen_potion.Is_Intelligence_pot:
        player.stats.intelligence += chosen_potion.Stat_Buff
        slow_print(f"You used {chosen_potion.name} and gained {chosen_potion.Stat_Buff} Intelligence!")
    elif chosen_potion.Is_speed_pot: # Assuming Is_speed_pot maps to agility
        player.stats.agility += chosen_potion.Stat_Buff
        slow_print(f"You used {chosen_potion.name} and gained {chosen_potion.Stat_Buff} Agility!")

    player.inventory.remove(chosen_potion)
    time.sleep(1)
    return True
def check_enemy(enemy):
    slow_print("\n--- Enemy Stats ---")
    time.sleep(1)
    slow_print(f"Name: {enemy.name}")
    time.sleep(1)
    slow_print(f"HP: {enemy.hp}/{enemy.max_hp()}")
    time.sleep(1)
    slow_print(enemy.base_stats.__check_stats__())
    time.sleep(1)
    slow_print(f"Weapon: {enemy.weapon.name}")
    time.sleep(1)

def player_turn(player, enemy):
    slow_print("\nYour turn!")
    time.sleep(1)
    slow_print(f"{enemy.name} HP: {enemy.hp}/{enemy.max_hp()}")
    time.sleep(1)
    slow_print("Actions: attack | run | potion | check")
    time.sleep(1)
    while True:
      action = input("> ").lower()

      if action == "attack":
          num_attacks_this_turn = 0
          miss_attack = False
          weapons_to_attack_with = []

          # Determine which weapons to attack with and how many times
          if player.weapon and player.weapon.dual_wield and player.second_weapon and player.second_weapon.dual_wield:
              slow_print(f"You dual-wield your {player.weapon.name} and {player.second_weapon.name} and attack twice!")
              time.sleep(1)
              weapons_to_attack_with.append(player.weapon)
              weapons_to_attack_with.append(player.second_weapon)
          elif player.weapon:
              weapons_to_attack_with.append(player.weapon)
              if player.weapon.dual_wield:
                  slow_print("You dual-wield your weapon and attack twice!")
                  time.sleep(1)
                  weapons_to_attack_with.append(player.weapon)
              elif player.weapon.light:
                  if random.random() < 0.5:
                      slow_print("Your light weapon allows you to move quickly and attack twice!")
                      time.sleep(1)
                      weapons_to_attack_with.append(player.weapon)
              elif player.weapon.heavy:
                  if random.random() < 0.5:
                      miss_attack = True
                      slow_print("Your heavy weapon makes you lose balance, and you miss your attack!")
                      time.sleep(1)

          if miss_attack:
              pass # Do nothing, effectively a miss
          else:
              for weapon_used in weapons_to_attack_with:
                  dmg = calculate_damage(player, weapon_used, enemy)
                  enemy.hp -= dmg
                  slow_print(f"You attack {enemy.name} with your {weapon_used.name} for {dmg} damage!")
                  time.sleep(1)

                  # Apply special effects after damage
                  if weapon_used.has_lifesteal:
                      heal_amount = weapon_used.lifesteal_base + (player.stats.constitution * weapon_used.lifesteal_per_con)
                      player.hp = min(player.max_hp, player.hp + int(heal_amount))
                      slow_print(f"You lifesteal {int(heal_amount)} HP!")
                      time.sleep(1)

                  if weapon_used.has_poison and random.random() < weapon_used.poison_chance:
                      enemy.is_poisoned = True
                      enemy.poison_remaining_turns = weapon_used.poison_duration
                      enemy.poison_damage_on_turn = int(dmg * weapon_used.poison_damage_multiplier)
                      slow_print(f"{enemy.name} is poisoned for {enemy.poison_remaining_turns} turns, taking {enemy.poison_damage_on_turn} poison damage per turn!")
                      time.sleep(1)

                  if weapon_used.has_stun and random.random() < weapon_used.stun_chance:
                      enemy.is_stunned = True
                      enemy.stun_remaining_turns = weapon_used.stun_duration
                      slow_print(f"{enemy.name} is stunned for {enemy.stun_remaining_turns} turn(s)!")
                      time.sleep(1)

                  if weapon_used.has_fear and random.random() < weapon_used.fear_chance:
                      enemy.is_feared = True
                      enemy.fear_remaining_turns = weapon_used.fear_duration
                      enemy.feared_damage_reduction = weapon_used.fear_damage_reduction
                      slow_print(f"{enemy.name} is feared for {enemy.fear_remaining_turns} turns, reducing their damage by {int(enemy.feared_damage_reduction * 100)}%!")
                      time.sleep(1)

                  if enemy.hp <= 0: # If enemy dies after a hit
                      break
          return "continue"


      elif action == "run":
          if player.stats.agility > enemy.base_stats.agility:
              slow_print("You successfully escaped!")
              time.sleep(1)
              return "escaped"

          else:
              slow_print("You failed to escape!")
              time.sleep(1)
              return "continue"


      elif action == "potion":
          used = use_potion(player)
          return "continue" if used else "continue"


      elif action == "check":
          check_enemy(enemy)
          return "continue"


      else:
          slow_print("Invalid action! Please try again")
          time.sleep(1)
          continue

def enemy_turn(enemy, player):
    slow_print(f"\n{enemy.name}'s turn!")

    # Apply poison damage at the start of the enemy's turn
    if enemy.is_poisoned and enemy.poison_remaining_turns > 0:
        poison_dmg = enemy.poison_damage_on_turn
        enemy.hp -= poison_dmg
        slow_print(f"{enemy.name} takes {poison_dmg} poison damage!")
        enemy.poison_remaining_turns -= 1
        if enemy.poison_remaining_turns <= 0:
            enemy.is_poisoned = False
            enemy.poison_damage_on_turn = 0
        time.sleep(1)
        if enemy.hp <= 0: # Enemy might die from poison
            return # Skip attack if dead

    # Check for stun effect
    if enemy.is_stunned and enemy.stun_remaining_turns > 0:
        slow_print(f"{enemy.name} is stunned and cannot attack!")
        enemy.stun_remaining_turns -= 1
        if enemy.stun_remaining_turns <= 0:
            enemy.is_stunned = False
        time.sleep(1)
        return # Skip attack

    # Decrement fear duration
    if enemy.is_feared and enemy.fear_remaining_turns > 0:
        enemy.fear_remaining_turns -= 1
        if enemy.fear_remaining_turns <= 0:
            enemy.is_feared = False
            enemy.feared_damage_reduction = 0.0
            slow_print(f"{enemy.name} is no longer feared.")
        time.sleep(1)

    dmg = calculate_damage(enemy, enemy.weapon, player)
    player.hp -= dmg
    slow_print(f"{enemy.name} hits you for {dmg} damage!")
def combat(player, enemy):
    global kill_count
    enemy.hp = enemy.max_hp()
    # Reset enemy status effects at the start of combat
    enemy.is_poisoned = False
    enemy.poison_remaining_turns = 0
    enemy.poison_damage_on_turn = 0
    enemy.is_stunned = False
    enemy.stun_remaining_turns = 0
    enemy.is_feared = False
    enemy.fear_remaining_turns = 0
    enemy.feared_damage_reduction = 0.0

    slow_print(f"\nA wild {enemy.name} appears!")
    time.sleep(1)
    slow_print(enemy.description)
    time.sleep(1)
    player_first = player.stats.agility >= enemy.base_stats.agility

    while player.hp > 0 and enemy.hp > 0:
        if player_first:
            result = player_turn(player, enemy)
            if result == "escaped":
                return True
            if enemy.hp <= 0:
                break
            enemy_turn(enemy, player)
        else:
            enemy_turn(enemy, player)
            if player.hp <= 0:
                break
            result = player_turn(player, enemy)
            if result == "escaped":
                return True

        slow_print(f"\n{player.name} HP: {player.hp}/{player.max_hp}")

    if player.hp > 0:
        slow_print(f"\nYou defeated the {enemy.name}!")
        time.sleep(1)
        player.xp += enemy.exp
        slow_print(f"You gained {enemy.exp} XP!")
        kill_count = kill_count+1
        slow_print(f"you have {kill_count} kills")
        return True # Player won
    else:
        time.sleep(1)
        slow_print("\nYou Die")
        time.sleep(1)
        slow_print(f"you got {kill_count} kills")
        return False # Player lost
```
